# <a id="format">3. Format</a>

Need to convert this into the following format: **Cue | Response | Count** <br>
Note that we're not currently differentiating between R1, R2, and R3. Ideally, R1 must have a higher weightage of sorts.

```python
#D.groupby(['cue'],as_index=False).head()
#D[['cue','R1']].value_counts()
D1 = [list(a) for a in list(D[['cue','R1']].values)]
D2 = [list(a) for a in list(D[['cue','R2']].values)]
D3 = [list(a) for a in list(D[['cue','R3']].values)]
D123 = D1+D2+D3
D123
```

```python
len(D123),len(D123)/3,len(D1) # (3684600, 1228200.0, 1228200)
```

```python
D123_dict = {}
for i in D123:
    try:
        D123_dict[tuple(i)] +=1
    except:
        D123_dict[tuple(i)] = 1
D123_dict
```

```python
len(D123_dict) # unique (cue, response) pairs: 1409196
```

```python
D123_cr = {} # dict of cue: [list of (response,count)]
for cr, count in D123_dict.items(): 
    try:
        D123_cr[cr[0]].append(tuple([cr[1],count]))
    except:
        D123_cr[cr[0]] = [tuple([cr[1],count])]
D123_cr
```

```python
len(D123_cr) # number of cues: 12282
cues = list(D123_cr.keys())
len(cues) # 12282
```

# <a id="explore">4. Explore</a>

```python
import sortedcontainers as sc
from collections import namedtuple

RespCount = namedtuple('RespCount', ['response', 'count'])

def sortkey(elem):
    #return -1*elem.count
    return -1*elem[1]

def distance(r): # takes a tuple (response, count)
        return RespCount(response=r[0], count=r[1])

def top_count(responses, n): # takes a list of tuples (response, count)
        topten = sc.SortedListWithKey(key=sortkey)
        for r in responses:
            #topten.add(distance(r))
            topten.add(r)
            if len(topten) > n:
                topten.pop()
        return topten
```

```python
def view_D123_cr(D123_cr, topn=10, min_count=0): # can add all restrictions here itself, eg. topn, min_count.
    while(True):
        pick = np.random.randint(0, len(D123_cr))
        cue = cues[pick]
        responses = D123_cr[cue]
        responses = [r for r in responses if r[1]>=min_count]
        if len(responses) > 0:
            break
    top = list(top_count(responses,topn))
    print(cue)
    for i in top:
        print(i)
```

```python
view_D123_cr(D123_cr, topn=20, min_count=20)
```

# <a id="subset">5. Subset</a>
Select a subset of these to serve as your own version of SWOW-NNNN dataset.<br>
Primarily pick restrictions on count.

```python
min_count = 20
multi_words = False # if False, disallow cues and responses with more than 1 word.
incorrect_words = False # if False, disallow cues and responses which are not English words.

cr_dict = {}
hm_cues = 0
hm_resps = 0
for cue,resps in D123_cr.items():
    if type(cue) != str:
        continue
    if multi_words == False:
        if ' ' in cue:
            continue
    # also check for incorrect
    for r in resps:
        if type(r[0]) != str:
            continue
        if multi_words == False:
            if ' ' in r[0]:
                continue
        # also check for incorrect
        if r[1]>=min_count:
            try:
                cr_dict[cue].append(r[0])
            except:
                cr_dict[cue] = [r[0]]
            hm_resps+=1
            
hm_cues = len(cr_dict) # no. of cues, eg. earlier it was 8500 in SWOW-8500.
print(hm_cues, hm_resps)
# for min_count = 5, allow incorrect, allow multiwords:     (12282, )
# for min_count = 5, allow incorrect, disallow multiwords:  (11897, 126269)
# for min_count = 10, allow incorrect, disallow multiwords: (11871,  54934)
# for min_count = 20, allow incorrect, disallow multiwords: (10572,  21057)
```
